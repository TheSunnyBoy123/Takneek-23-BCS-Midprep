```python
import pandas as pd
```

```python
filename = "/Users/sunrajpatel/Desktop/code/takneek/do_i_know_you/S08_question_answer_pairs.txt"
df = pd.read_csv(filename, sep='\t', lineterminator='\n')
```

```python
df
```

```python
df = df[["ArticleTitle", "Question", "Answer"]]
df
```

```python
find_name = "Isaac_Newton"

new_df = df[df["ArticleTitle"] == find_name]

new_df
```

```python
new_df = new_df[["Question", "Answer"]]
new_df.dropna()
new_df.drop_duplicates()

new_df
```

```python
final_dict = dict()

#for each row in new_df, store question in question var and answer in answer var
for index, row in new_df.iterrows():
    question = row["Question"]
    answer = row["Answer"]
    final_dict[question] = answer

final_dict
```

```python
str = ""
for key, value in final_dict.items():
    str += '{ "instruction": "' + key + '",' + '\n' + '"input": "",' + '\n' + '"output": "' + value + '"},'  
```

```python
str
```

```python
#write str to output_file.json
output_file = "output_file.json"
with open(output_file, "w") as file:
    file.write(str)
    
```
